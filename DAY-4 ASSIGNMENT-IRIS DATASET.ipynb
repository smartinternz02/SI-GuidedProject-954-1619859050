```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.pairplot(iris)
```

```python
sns.scatterplot(x = "sepal_length",y = "petal_length",data=iris)
```

```python
sns.scatterplot(x = "sepal_width",y = "petal_width",data=iris)
```

```python
sns.lineplot(x = "sepal_length",y = "petal_length",data=iris)
```

As sepal length increases petal length also increases

```python
sns.lineplot(x = "sepal_width",y = "petal_width",data=iris)
```

In between 3.0 to 3.5(sepal width) petal width is high

```python
sns.barplot(x="species",y="petal_length",data=iris)
```

```python
sns.barplot(x="species",y="petal_width",data=iris)
```

```python
sns.barplot(x="species",y="sepal_length",data=iris)
```

```python
sns.barplot(x="species",y="sepal_width",data=iris)
```

From above 4 barplots we can say that length of petal,sepal and width of sepal is more in virginica.In setosa sepal width is more.
